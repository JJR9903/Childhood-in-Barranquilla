Se importan y limpian los datos en dos loops, 1 para los resultados del icfes antes del cambio del puntaje global en el 2014-2, y el 2 para los resultados despues de este cambio

```python
for key in Icfes_1.keys():
    Icfes_1[key]["df"] =  pd.read_csv(Icfes_1[key]["dir"],delimiter=';',
                      usecols= ["COLE_DEPTO_UBICACION","COLE_COD_DEPTO_UBICACION","COLE_MCPIO_UBICACION","COLE_COD_MCPIO_UBICACION",'PUNT_MATEMATICAS','PUNT_INGLES','PUNT_CIENCIAS_SOCIALES','PUNT_BIOLOGIA','PUNT_FILOSOFIA','PUNT_FISICA','PUNT_QUIMICA','PUNT_LENGUAJE','COLE_NATURALEZA'],
                     dtype = {"COLE_DEPTO_UBICACION":"str","COLE_COD_DEPTO_UBICACION":"str","COLE_MCPIO_UBICACION":"str","COLE_COD_MCPIO_UBICACION":"str",'PUNT_MATEMATICAS':"str",'PUNT_INGLES':"str",'PUNT_CIENCIAS_SOCIALES':"str",'PUNT_BIOLOGIA':"str",'PUNT_FILOSOFIA':"str",'PUNT_FISICA':"str",'PUNT_QUIMICA':"str",'PUNT_LENGUAJE':"str",'COLE_NATURALEZA':"str"},low_memory=False)
    Icfes_1[key]["df"].columns = list(map(str.upper,Icfes_1[key]["df"].columns))
    Icfes_1[key]["df"].rename(columns = {'COLE_DEPTO_UBICACION':'DEPTO','COLE_MCPIO_UBICACION':'MUNICIPIO',"COLE_COD_DEPTO_UBICACION":'COD_DEPTO','COLE_COD_MCPIO_UBICACION':'COD_MUN'}, inplace = True)
    Icfes_1[key]["df"] = Icfes_1[key]["df"].assign(
        COD_DEPTO = pd.to_numeric(Icfes_1[key]["df"].COD_DEPTO),
        COD_MUN = pd.to_numeric(Icfes_1[key]["df"].COD_MUN),
        PUNT_MATEMATICAS = preprocessing.scale(pd.to_numeric(Icfes_1[key]["df"].PUNT_MATEMATICAS.str.replace(',', '.'))),
        PUNT_LENGUAJE = preprocessing.scale(pd.to_numeric(Icfes_1[key]["df"].PUNT_LENGUAJE.str.replace(',', '.'))),
        PUNT_INGLES = preprocessing.scale(pd.to_numeric(Icfes_1[key]["df"].PUNT_INGLES.str.replace(',', '.'))),
        PUNT_CIENCIAS_SOCIALES = preprocessing.scale(pd.to_numeric(Icfes_1[key]["df"].PUNT_CIENCIAS_SOCIALES.str.replace(',', '.'))),
        PUNT_FISICA = preprocessing.scale(pd.to_numeric(Icfes_1[key]["df"].PUNT_FISICA.str.replace(',', '.'))),
        PUNT_QUIMICA = preprocessing.scale(pd.to_numeric(Icfes_1[key]["df"].PUNT_QUIMICA.str.replace(',', '.'))),
        PUNT_BIOLOGIA = preprocessing.scale(pd.to_numeric(Icfes_1[key]["df"].PUNT_BIOLOGIA.str.replace(',', '.'))),
        PUNT_FILOSOFIA = preprocessing.scale(pd.to_numeric(Icfes_1[key]["df"].PUNT_FILOSOFIA.str.replace(',', '.'))),
    )
    Icfes_1[key]["df"] = Icfes_1[key]["df"].assign(PUNT_GLOBAL = (Icfes_1[key]["df"].PUNT_QUIMICA+Icfes_1[key]["df"].PUNT_BIOLOGIA+Icfes_1[key]["df"].PUNT_FISICA+(Icfes_1[key]["df"].PUNT_CIENCIAS_SOCIALES*2)+Icfes_1[key]["df"].PUNT_FILOSOFIA+(Icfes_1[key]["df"].PUNT_LENGUAJE*3)+(Icfes_1[key]["df"].PUNT_MATEMATICAS*3)+Icfes_1[key]["df"].PUNT_INGLES)/13)
    Icfes_1[key]["df"] = Icfes_1[key]["df"].loc[(Icfes_1[key]["df"]["COD_MUN"]==11001) | (Icfes_1[key]["df"]["COD_MUN"]==76001) | (Icfes_1[key]["df"]["COD_MUN"]==5001) | (Icfes_1[key]["df"]["COD_MUN"]==8001) | (Icfes_1[key]["df"]["COD_MUN"]==68001) | (Icfes_1[key]["df"]["COD_MUN"]==17001) | (Icfes_1[key]["df"]["COD_MUN"]==66001) | (Icfes_1[key]["df"]["COD_MUN"]==54001) | (Icfes_1[key]["df"]["COD_MUN"]==52001) | (Icfes_1[key]["df"]["COD_MUN"]==73001) | (Icfes_1[key]["df"]["COD_MUN"]==23001) | (Icfes_1[key]["df"]["COD_MUN"]==13001) | (Icfes_1[key]["df"]["COD_MUN"]==50001)]
    Icfes_1[key]["df"] = Icfes_1[key]["df"].loc[(Icfes_1[key]["df"]["COLE_NATURALEZA"]=="OFICIAL")] 
    Icfes_1[key]["df"] = Icfes_1[key]["df"][["DEPTO","COD_DEPTO","MUNICIPIO","COD_MUN","PUNT_GLOBAL","PUNT_INGLES"]]
    print(key)

```

```python
for key in Icfes_2.keys():
    if key=="p2015_1":
        Icfes_2[key]["df"] =  pd.read_csv(Icfes_2[key]["dir"],delimiter=',',
                      usecols= ["cole_depto_ubicacion","cole_cod_depto_ubicacion","cole_mcpio_ubicacion","cole_cod_mcpio_ubicacion",'punt_global','punt_ingles','cole_naturaleza'],
                     dtype = {"cole_depto_ubicacion":"str","cole_cod_depto_ubicacion":"str","cole_mcpio_ubicacion":"str","cole_cod_mcpio_ubicacion":"str",'punt_global':"str",'punt_ingles':"str",'cole_naturaleza':"str"},low_memory=False)
    else:
        Icfes_2[key]["df"] =  pd.read_csv(Icfes_2[key]["dir"],delimiter=';',
                      usecols= ["COLE_DEPTO_UBICACION","COLE_COD_DEPTO_UBICACION","COLE_MCPIO_UBICACION","COLE_COD_MCPIO_UBICACION",'PUNT_GLOBAL','PUNT_INGLES','COLE_NATURALEZA'],
                     dtype = {"COLE_DEPTO_UBICACION":"str","COLE_COD_DEPTO_UBICACION":"str","COLE_MCPIO_UBICACION":"str","COLE_COD_MCPIO_UBICACION":"str",'PUNT_GLOBAL':"str",'PUNT_INGLES':"str",'COLE_NATURALEZA':"str"},low_memory=False)

    Icfes_2[key]["df"].columns = list(map(str.upper,Icfes_2[key]["df"].columns))
    Icfes_2[key]["df"].rename(columns = {'COLE_DEPTO_UBICACION':'DEPTO','COLE_MCPIO_UBICACION':'MUNICIPIO',"COLE_COD_DEPTO_UBICACION":'COD_DEPTO','COLE_COD_MCPIO_UBICACION':'COD_MUN'}, inplace = True)
    Icfes_2[key]["df"] = Icfes_2[key]["df"].assign(
        COD_DEPTO = pd.to_numeric(Icfes_2[key]["df"].COD_DEPTO),
        COD_MUN = pd.to_numeric(Icfes_2[key]["df"].COD_MUN),
        PUNT_GLOBAL = preprocessing.scale(pd.to_numeric(Icfes_2[key]["df"].PUNT_GLOBAL.str.replace(',', '.'))),
        PUNT_INGLES = preprocessing.scale(pd.to_numeric(Icfes_2[key]["df"].PUNT_INGLES.str.replace(',', '.')))
    )
    Icfes_2[key]["df"] = Icfes_2[key]["df"].loc[(Icfes_2[key]["df"]["COD_MUN"]==11001) | (Icfes_2[key]["df"]["COD_MUN"]==76001) | (Icfes_2[key]["df"]["COD_MUN"]==5001) | (Icfes_2[key]["df"]["COD_MUN"]==8001) | (Icfes_2[key]["df"]["COD_MUN"]==68001) | (Icfes_2[key]["df"]["COD_MUN"]==17001) | (Icfes_2[key]["df"]["COD_MUN"]==66001) | (Icfes_2[key]["df"]["COD_MUN"]==54001) | (Icfes_2[key]["df"]["COD_MUN"]==52001) | (Icfes_2[key]["df"]["COD_MUN"]==73001) | (Icfes_2[key]["df"]["COD_MUN"]==23001) | (Icfes_2[key]["df"]["COD_MUN"]==13001) | (Icfes_2[key]["df"]["COD_MUN"]==50001)]
    Icfes_2[key]["df"] = Icfes_2[key]["df"].loc[(Icfes_2[key]["df"]["COLE_NATURALEZA"]=="OFICIAL")] 
    Icfes_2[key]["df"] = Icfes_2[key]["df"][["DEPTO","COD_DEPTO","MUNICIPIO","COD_MUN","PUNT_GLOBAL","PUNT_INGLES"]]
    print(key)
```

Se juntan los datos para cada año, y se agrupan por ciudades

```python
for p in ['p2011_1','p2011_2','p2012_1','p2012_2','p2013_1','p2013_2','p2014_1']:
    Icfes_1[p]["df"]=Icfes_1[p]["df"].replace(['BOGOTÁ, D.C.','MEDELLIN','MONTERIA','CARTAGENA'],['MEDELLÍN','MONTERÍA','CARTAGENA DE INDIAS','BOGOTÁ D.C.'])

for p in ['p2014_2','p2015_1','p2015_2','p2016_1','p2016_2','p2017_1','p2017_2','p2018_1','p2018_2','p2019','p2020_1','p2020_2','p2021_1','p2021_2']:
    Icfes_2[p]["df"]=Icfes_2[p]["df"].replace(['BOGOTÁ, D.C.','MEDELLIN','MONTERIA','CARTAGENA'],['MEDELLÍN','MONTERÍA','CARTAGENA DE INDIAS','BOGOTÁ D.C.'])
```

```python
p2011=pd.concat([Icfes_1["p2011_1"]["df"],Icfes_1["p2011_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2012=pd.concat([Icfes_1["p2012_1"]["df"],Icfes_1["p2012_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2013=pd.concat([Icfes_1["p2013_1"]["df"],Icfes_1["p2013_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2014=pd.concat([Icfes_1["p2014_1"]["df"],Icfes_2["p2014_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2015=pd.concat([Icfes_2["p2015_1"]["df"],Icfes_2["p2015_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2016=pd.concat([Icfes_2["p2016_1"]["df"],Icfes_2["p2016_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2017=pd.concat([Icfes_2["p2017_1"]["df"],Icfes_2["p2017_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2018=pd.concat([Icfes_2["p2018_1"]["df"],Icfes_2["p2018_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2019=Icfes_2["p2019"]["df"][['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2020=pd.concat([Icfes_2["p2020_1"]["df"],Icfes_2["p2020_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]
p2021=pd.concat([Icfes_2["p2021_1"]["df"],Icfes_2["p2021_2"]["df"]])[['MUNICIPIO','PUNT_GLOBAL','PUNT_INGLES']]

p2011=p2011.groupby('MUNICIPIO').mean()
p2012=p2012.groupby('MUNICIPIO').mean()
p2013=p2013.groupby('MUNICIPIO').mean()
p2014=p2014.groupby('MUNICIPIO').mean()
p2015=p2015.groupby('MUNICIPIO').mean()
p2016=p2016.groupby('MUNICIPIO').mean()
p2017=p2017.groupby('MUNICIPIO').mean()
p2018=p2018.groupby('MUNICIPIO').mean()
p2019=p2019.groupby('MUNICIPIO').mean()
p2020=p2020.groupby('MUNICIPIO').mean()
p2021=p2021.groupby('MUNICIPIO').mean()

p2011.loc["13 CIUDADES PRINCIPALES"]=p2011.mean()
p2012.loc["13 CIUDADES PRINCIPALES"]=p2012.mean()
p2013.loc["13 CIUDADES PRINCIPALES"]=p2013.mean()
p2014.loc["13 CIUDADES PRINCIPALES"]=p2014.mean()
p2015.loc["13 CIUDADES PRINCIPALES"]=p2015.mean()
p2016.loc["13 CIUDADES PRINCIPALES"]=p2016.mean()
p2017.loc["13 CIUDADES PRINCIPALES"]=p2017.mean()
p2018.loc["13 CIUDADES PRINCIPALES"]=p2018.mean()
p2019.loc["13 CIUDADES PRINCIPALES"]=p2019.mean()
p2020.loc["13 CIUDADES PRINCIPALES"]=p2020.mean()
p2021.loc["13 CIUDADES PRINCIPALES"]=p2021.mean()

p2011["AÑO"]=2011
p2012["AÑO"]=2012
p2013["AÑO"]=2013
p2014["AÑO"]=2014
p2015["AÑO"]=2015
p2016["AÑO"]=2016
p2017["AÑO"]=2017
p2018["AÑO"]=2018
p2019["AÑO"]=2019
p2020["AÑO"]=2020
p2021["AÑO"]=2021

ICFES = pd.concat([p2011,p2012,p2013,p2014,p2015,p2016,p2017,p2018,p2019,p2020,p2021])
ICFES['MUNICIPIO']=ICFES.index
ICFES=ICFES.loc[(ICFES['MUNICIPIO']=='BARRANQUILLA') | (ICFES['MUNICIPIO']=='BOGOTÁ D.C.') | (ICFES['MUNICIPIO']=='BOGOTÁ, D.C.') | (ICFES['MUNICIPIO']=='BUCARAMANGA') | (ICFES['MUNICIPIO']=='CALI') | (ICFES['MUNICIPIO']=='CARTAGENA DE INDIAS') | (ICFES['MUNICIPIO']=='MEDELLÍN') | (ICFES['MUNICIPIO']=='MONTERÍA') | (ICFES['MUNICIPIO']=='13 CIUDADES PRINCIPALES')]
```
